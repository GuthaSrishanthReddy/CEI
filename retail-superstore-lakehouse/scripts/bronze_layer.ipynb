```python
raw_path = "/Volumes/workspace/default/retail_data/raw/"
bronze_path = "/Volumes/workspace/default/retail_data/bronze/"
```

```python
#Writing the raw data into bronze layer
for name in ["customers", "products", "orders", "payments"]:
    df = spark.read.option("header", True).option("inferSchema", True).csv(raw_path + name)
    df.write.mode("overwrite").option("header", True).csv(bronze_path + name)
```
