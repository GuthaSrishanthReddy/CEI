```python
import pandas as pd
```

```python
df = pd.read_csv("Combined_dataset.csv")
df.head(3)
```

```python
# explore data
print("Shape of the data:", df.shape)
print('====================================================================================================================')
print("Columns in the data:", df.columns)
print('====================================================================================================================')
print("Data types of each column:\n", df.dtypes)
print('====================================================================================================================')
print("Number of missing values in each column:\n", df.isnull().sum())
```

```python
# Handling NaN values

df1 = df.copy()

df1["what_customers_said"] = df1["what_customers_said"].fillna("No comments")
df1["seller_name"] = df1["seller_name"].fillna("Unknown")
df1["seller_information"] = df1["seller_information"].fillna("Unknown")
df1["discount"] = df1["discount"].fillna(0)
df1["videos"] = df1["videos"].fillna("No Video")
df1["variations"] = df1["variations"].fillna("No Variants")
#using another field to know if there are any variations or not, as it would be easier to filter the products with variations if needed!
df1["has_variations"] = df1["variations"].notna()

df1.head(3)

#There are no columns in the dataset that can be wither filled with any statistic or any fillna technique, they're all categorical and textual
#so it's better to fill them with a placeholder value that indicates the absence of the data.
```

```python
# basic operations: filter rows and select columns
selected_columns = df1[["product_id", "title", "category", "rating", "initial_price", "discount"]]
high_rated = selected_columns[(df1["rating"] > 4) & (df1["ratings_count"] > 70)]
high_rated.head(3)
```

```python
# drop duplicates
df1 = df1.drop_duplicates(subset=["product_id"])
df1.shape
```

```python
# Derived column
df1["calculated_final_price"] = df1["initial_price"] - (df1["initial_price"] * df1["discount"] / 100)
df1["calculated_final_price"] = df1["calculated_final_price"]

df1.head(3)
```

```python
# save cleaned combined dataset
df1.to_csv("cleaned_Combined_dataset.csv", index=False)
df1.head(3)
```
