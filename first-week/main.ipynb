```python
import pandas as pd
```

```python
df = pd.read_csv("sports-shoes.csv")
df.head(3)
```

```python
#explore data
print("Shape of the data:", df.shape)
print('====================================================================================================================')
print("Columns in the data:", df.columns)
print('====================================================================================================================')
print("Data types of each column:\n", df.dtypes)
print('====================================================================================================================')
print("Number of missing values in each column:\n", df.isnull().sum())
```

```python
#Handling Nan values
df1 = df.drop(columns=["what_customers_said"])
df1["seller_name"] = df1["seller_name"].fillna(value="Unknown")
df1["seller_information"] = df1["seller_information"].fillna(value="Unknown")
df1["discount"] = df1["discount"].fillna(value=0)
df1["videos"] = df1["videos"].fillna(value="https://youtu.be/dQw4w9WgXcQ?t=0")
df1["variations"] = df1["variations"].fillna(value="No Variants")
df1.head(3)
```

```python
#basic operations (filter rows, select columns)
df1[df1["rating"] > 3].head(3)
df1[(df1["rating"] > 4) & (df1["ratings_count"] > 70)][["product_id", "title"]].head(3)
```

```python
#drop duplicates
df1 = df1.drop_duplicates(subset=["product_id"])
df1.shape
```

```python
#Derived column 
df1["final_price"] = df1["initial_price"] - (df1["initial_price"] * df1["discount"] / 100)
df1.head(3)
```

```python
df = df1
df
df.to_csv("cleaned_sports-shoes.csv", index=False)
```
